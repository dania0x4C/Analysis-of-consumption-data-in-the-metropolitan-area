# 「2024년 통계데이터 활용대회」데이터분석 보고서

## 제목: 지역별 카드 소비 매출 분석 및 정책 제안

## 1. 배경
최근 물가 상승과 소득 및 지가의 격차로 인한 지역 간 경제적 불균형이 심화되고 있다. 이는 단순히 서울과 다른 지역 간의 문제뿐만 아니라 서울 내부에서도 경제적 격차가 점점 커지고 있는 상황이다. 이러한 지역 간 경제위기 문제에 대한 심각성을 인지하고 해결 방안을 모색하는 것이 필요하다.
이 문제를 해결하기 위한 대표적인 정책 중 하나로 '지역화폐'가 도입되었다.
지역화폐는 소상공인과 자영업자 지원을 통해 지역공동체 강화 및 지역경제를 활성화하고 지역 내 소비를 촉진하는 데 의미 있는 효과를 가지고 있다. 그러나 현재 지역화폐 정책의 예산이 줄어들고 있으며, 실질적으로 효과를 보지 못 하는 지자체도 있다.



[자료 | 코드포코리아_지역화폐_통계자료]

전국 243개 광역 및 기초 지자체 중 지난해 지역화폐 발행액을 전년 대비 확대한 지자체는 57개뿐이었다. 전년 수준으로 유지한 곳은 7곳으로, 전체 지자체 중 25.1%에 불과했다. 나머지 75%에 달하는 지자체는 지역화폐 예산을 축소하거나 발행을 중지했다. 이처럼 현재 진행 중인 지역화폐 정책은 지속 가능성에 한계가 있다는 걸 알 수 있다.
또한, 현재 지역화폐는 모든 업종에 일정 할인 혜택을 적용하고 있지만, 이는 각 지역의 산업 특성과 지역 소비자의 니즈를 충분히 반영하지 못하고 있다. 각 지역에 맞는 특정 업종에 대한 지원이 필요하다.
따라서, 본 보고서에서는 지역별 카드 소비 매출 데이터를 분석하여 각 지역에 맞는 맞춤형 정책을 제안하고자 한다. 이를 통해 지역 간 경제적 격차를 줄이고 보다 효과적인 지역 경제 활성화를 도모할 수 있을 것이다.

## 2. 데이터 분석

### 데이터 선정(사용한 데이터 및 이유 등)

1. (신한카드)내국인 카드소비 데이터
   이 연구에서는 SDC 통계 데이터 센터에서 제공하는 민간 자료인 내국인 국내 카드 소비(신한카드) 데이터를 주요 자료로 활용했다. 이 자료는 2019년 10월부터 2023년 12월까지의 시계열 정보를 포함한다. 기준년월, 고객 주소지, 성별, 연령별, 직업 구분, 가구 유형별, 업종별 소비금액과 건수를 제공한다. 이 자료는 다양한 변수를 포괄적으로 분석할 수 있는 소비 데이터 자료로서 연구에 적합하다고 판단하여 선택했다.
   연구에서는 최신 데이터인 2023년 1월부터 12월까지의 정보를 모두 사용했다. 특히, 지역별 소비 패턴이 계절적 요소에 영향을 받을 가능성을 고려하여, 연간 데이터를 분기별로 분류하여 세부적인 분석에 활용했다.
2. json 지역 경계 데이터
   지도 시각화를 위해 통계청 통계 지리 정보 서비스에서 제공하는 행정동 경계 파일을 기반으로 한 대한민국 행정동 경계 데이터인 GEOJSON 데이터를 활용했다. 이 데이터는 지리적 정보를 직관적으로 시각화하는데 활용했다.

### 데이터 전처리
  이 연구에서는 내국인 국내 카드 소비(신한카드) 데이터 중에서 기준년월, 고객주소지 광역시도명, 고객주소지 시군구명, 연령대별, 업종 대분류명, 업종 소분류명, 카드 이용 금액 변수를 활용했다.
  기준년월 변수를 통해 2023년 데이터를 4분기로 나누어 분석하였고, 분기별 소비 패턴을 파악하는 데 사용했다.
  고객주소지 광역시도명과 시군구명 변수를 사용하여 지역을 분류했다. 전체 데이터를 사용하기에는 그 수가 너무 방대하므로, 수도권에 대한민국 인구의 약 50%가 밀집해 있는 도시 집중 현상을 고려하여 분석 범위를 전국이 아닌 서울특별시, 경기도, 인천광역시로 한정했다.
  연령대는 10대, 20대, 30대, 40대, 50대, 60대 이상으로 분류하여 각 연령대의 소비 성향을 파악하는 데 활용했다.



[표 | 분석에 사용된 업종분류 종류]

  위 표는 사용데이터의 업종대분류, 업종소분류 열을 기준으로 새로운 업종분류의 종류로 정리한 것이다.
  기존 업종대분류의 ‘온라인’은 지역 소비를 필요로 하는 본 보고서의 취지에 맞지 않아 제거하였다.
  또한 기존 ‘유흥’ 분류에 포함되어 있던 ‘pc방/비디오방/전화방’과 ‘노래방을 ‘여가’ 분류로 옮겼다. 이는 pc방과 노래방을 소비자의 여가로 위한 것으로 간주하여 데이터 수가 많지 않은 유흥을 여가로 포함시켰다.
  ‘요식’에 포함되어 있던 ‘커피/제과’는 한국의 방대한 커피점 수를 고려하여 따로 새로운 분류명을 생성하였다. 나머지 요식 종류는 기존 ‘요식’의 이름으로 종류하였다.
  또한 따로 작성되어있던 ‘인테리어’ 분류의 ‘인테리어’,‘가전’,‘가구’는 그 데이터 수가 많지 않음을 고려하여 ‘생활/서비스’ 분류에 포함시켰다.
  카드 이용 금액은 (시군구별 업종별 카드 이용 금액 / 광역시별 업종별 카드 이용 금액)으로 계산하여 새로운 비율로 나타내어 사용했다. 이 비율을 통해 시군구별 소비 패턴을 비교하는 데 활용했다.

### EDA (탐색적 데이터 분석)



[그래프| 연령대별 업종 카드 사용량]

  위 그래프는 연령별 업종의 카드 사용량을 나타낸 것으로 새로 분류한 ‘연령대\_통합’을 기준으로 나누었다.
  이 그래프를 통해 40대의 카드 사용량이 제일 많은 것을 확인할 수 있었으며, 업종분류 중에서는 ‘유통’의 카드 사용량이 제일 많은 것으로 나타났다.
  업종분류 중 ‘유통’에 속한 업종소분류는 마트, 편의점, 백화점 등 실생활과 밀접한 관련이 있는 종류들이기 때문에 그 사용량이 많을 수밖에 없을 것으로 보인다.
  특징적으로, 20대와 60대 이상을 비교했을 때, 20대는 유통 업종 비율이 41.3%, 요식 업종 비율이 27.2%로 그 두 업종의 비율이 0.66이다. 반면 60대 이상은 유통 업종 비율이 41.4%, 요식 업종 비율이 18.8%로 그 비율이 0.45이다.
  또한 유통과 의료에 있어서 20대는 의료 업종 비율이 5.3%로 유통 업종에 대한 비율이 0.13인데 반해, 60대 이상은 의료 업종 비율이 14.6%로 유통 업종에 대한 비율이 0.35이다.
  이를 통해 20대는 요식 업종의 비율이 60대 이상에 비해 상대적으로 높고, 의료 업종의 비율이 상대적으로 낮다는 것을 알 수 있다.
  다시 말해, 세대별로 선호하는 업종의 차이를 명확하게 확인할 수 있으며, 이는 지역별 연령대에 따라 그 지역의 소비 업종에 영향을 미칠 것이라는 결론을 도출할 수 있다.



[그래프| 지역별 1등 카드 사용량 연령대_서울] [그래프| 지역별 1등 카드 사용량 연령대_경기,인천]

위 그래프는 지역별로 카드 사용량이 가장 많은 연령대를 표시한 그래프이다.
전체 대비 카드 사용량이 많은 30대, 40대가 대부분의 지역에서 특징을 두드러지게 나타나는 것을 알 수 있다. 이에 반해 특징적으로 서울의 마포구, 관악구에는 20대가 특징을 나타내는 것을 알 수 있고, 경기 인천의 외각 지역은 50대, 60대 이상이 특징을 나타내는 것을 알 수 있다.
일부 지역을 제외한 지역은 공통적으로 30대, 40대가 대부분을 차지하고 연령대 분포의 차이가 크지 않기 때문에 연령대별로 지역 정책에 차이를 두는 것은 유의미하지 않다고 판단했다. 이에 따라 지역별 업종 분석에 초점을 두어 정책 지원을 제안하려고 한다.

 다음은 2023년 분기별 지역별 타 업종 대비 가장 높은 카드사용 비율을 보인 업종을 그래프로 나타낸 것으로, 시각적 명확성을 위해 서울과 경기, 인천으로 나누어 표시하였다.
기본적으로 요식과 유통이 대부분의 지역에 많은 특성을 보일 수 밖에 없으므로, 위 그래프는 다른 지역과 비교했을 때 드러나는 업종 특성을 나타낸 것이다. 이를 기반으로 기존 요식과 유통에 더불어 각 지역별 업종을 추가적으로 정책 지원하는 것이 필요하다.
지역별 업종 카드 사용량을 나타냈을 때 서울의 대부분 지역이 경기, 인천에 비해 요식과 커피/제과 업종이 뚜렷한 특징을 보인다. 반면에 경기, 인천은 미용, 차량, 생활/서비스 등 다양한 업종의 카드 사용량이 두각을 나타내며 이는 서울 지역에서는 잘 보이지 않는 특성이라는 것을 알 수 있다.



[그래프| 분기별 변화를 보인 지역별 업종 변화추이]

이 그래프는 분기별로 1등 카드 사용량 업종이 빈번하게 변화한(3회 이상) 지역을 나타낸 것이다. 강남구, 서초구, 고양시, 여주군, 의정부시는 최근 2023년 안에 카드 사용량 주요 업종이 3번 이상 변화했다. 이는 특정 업종이 두각을 드러내기보다는 여러 업종에 고르게 카드 사용량이 발생하고 있다는 것을 의미한다. 그러므로 이 지역들은 각 업종별로 균형 잡힌 지원이 필요하다.

### 3. 분석 결과 및 해석
   지역 간의 경제적 불균형이 심화되고 있는 만큼, 각 지역의 소비 특성 또한 다르게 나타난다. 이 지역 간 특성의 차이를 이해하지 않고 동일한 지원을 제공할 경우, 정책의 효율성이 저하되고 낭비가 발생할 수 있다. 이는 대부분의 지역에서 기존 지역화폐 정책의 사용량이 줄어드는 가장 큰 이유라고 할 수 있다. 따라서 각 지역 간 소비 특성 차이를 이해하고, 지역별 맞춤 정책이 추가로 필요하다.
   기본적으로 유통과 요식이 우리 일상에서 대부분의 소비를 차지하고 있으므로 혜택에 집중해야 할 업종이다. 이와 더불어 각 지역의 소비 특성을 고려하여 타지역보다 소비가 많이 발생하는 업종에 추가 할인 혜택을 지원하는 것이 효과적일 것이다. 각 업종별로 다양한 할인율을 적용하여 소비자들의 정책 이용을 촉진시킬 수 있으며, 이는 결국 경제 순환에 기여하고 정책의 효율성을 높이는 데 도움이 될 것이다.
   또한 대부분의 지역에서 30대와 40대가 높은 카드 사용량을 보인 점을 고려하여 이들의 특성을 반영하는 것도 중요하다. 예를 들어, 직장인들이 많이 이용하는 점심 식사, 커피숍, 헬스장 등에서 지역화폐 사용 시 특별 할인을 제공하면 정책 활용의 기대를 높일 수 있을 것이다.
   다시 말해, 각 지역의 소비 특성은 유사하지만 차이점이 존재하며 이러한 특징들은 시간이 지남에 따라 지속적으로 변화할 것이다. 그렇기 때문에 각 지역의 특성을 이해하기 위해 상시 모니터링과 같은 지속적인 노력이 필요하다. 이를 통해 급변하는 소비 패턴에 신속히 대응할 수 있고, 지속 가능한 경제 발전을 지원할 수 있다.
